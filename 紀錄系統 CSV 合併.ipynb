```python
import os
import pandas as pd
```

```python
folder_path = r'C:/Users/ASUS/OneDrive/Documents/中信兄弟/Season_data/2024'  #紀錄系統
#folder_path = r'C:/Users/ASUS/OneDrive/Documents/中信兄弟/Trackman/2022/一軍'   #Trackman
file_extension = '.csv'

csv_files = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith(file_extension)]

columns_list = []

for file in csv_files:
    try:
        data = pd.read_csv(file)
        if not data.empty:
            columns_list.append(set(data.columns))
    except pd.errors.EmptyDataError:
        print(f"文件 '{file}' 是空的，跳過檢查。")

if columns_list:
    if all(columns_list[0] == columns for columns in columns_list):
        print("所有文件的欄位相同。")
    else:
        print("文件的欄位不同。")
else:
    print("沒有找到任何包含數據的文件。")
```

## 1.不保留 Trackman資料合併

```python
# List all CSV files in the directory
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

# Read the first file to get the column structure
first_file_path = os.path.join(folder_path, csv_files[0])
first_data = pd.read_csv(first_file_path)
#first_data.fillna(0, inplace=True)

# Initialize the DataFrame with the columns of the first file
all_data = pd.DataFrame(columns=first_data.columns)

# Process each CSV file
for file in csv_files:
    file_path = os.path.join(folder_path, file)
    data = pd.read_csv(file_path)
    #data.fillna(0, inplace=True)
    
    # Select only the columns that exist in the first file
    data = data.reindex(columns=first_data.columns, fill_value=0)
    
    # Concatenate the data
    all_data = pd.concat([all_data, data], ignore_index=True, sort=False)

# Display information about the concatenated DataFrame
all_data.info()
```

## 2.保留Trackman欄位合併

```python
folder_path = r'C:/Users/ASUS/OneDrive/Documents/中信兄弟/Season_data/2024'  # 紀錄系統
#folder_path = r'C:/Users/ASUS/OneDrive/Documents/中信兄弟/Trackman/2022/一軍'   # Trackman
file_extension = '.csv'

# 找到資料夾中的所有 CSV 檔案
csv_files = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith(file_extension)]

# 用來存儲合併後的所有數據
all_data = pd.DataFrame()

# 處理每個 CSV 檔案
for file in csv_files:
    try:
        data = pd.read_csv(file)
        
        # 如果檔案非空，則合併
        if not data.empty:
            # 合併數據，保留所有不同的欄位，缺失值填充為 NaN
            all_data = pd.concat([all_data, data], ignore_index=True, sort=False)
    except pd.errors.EmptyDataError:
        print(f"文件 '{file}' 是空的，跳過檢查。")

# 顯示合併後的數據資訊
all_data.info()
```
