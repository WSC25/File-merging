```python
import pandas as pd
import os
```

```python
import chardet

def detect_encoding(file_path):
    with open(file_path, 'rb') as f:
        result = chardet.detect(f.read())
    return result['encoding']

folder_path = r'C:\Users\ASUS\OneDrive\Documents\中信兄弟\Trackman\2024\一軍'
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]


first_file_path = os.path.join(folder_path, csv_files[0])
encoding = detect_encoding(first_file_path)
first_data = pd.read_csv(first_file_path, encoding=encoding)
first_data.fillna(0, inplace=True)

all_data = pd.DataFrame(columns=first_data.columns)

for file in csv_files:
    file_path = os.path.join(folder_path, file)
    encoding = detect_encoding(file_path)
    data = pd.read_csv(file_path, encoding=encoding)
    all_data = pd.concat([all_data, data], ignore_index=False, sort=False)

all_data.drop(columns=['PitchNo'], inplace=True)
```
